- 라디오 버튼 예시 코드: 첫 요소가 Default로 선택됨
    - 조건문 안에 코드 작성하여 해당하는 투자 성향에 따라 다른 포트폴리오 제공?

```python
    selected_item = st.radio("Radio Part", ('안정형', '중립형', '수익형'))

    if selected_item == "안정형":
        st.write("안정형!")
    elif selected_item == "중립형":
        st.write("중립형!")
    elif selected_item == "수익형":
        st.write("수익형!")
```
<br>

- 선택 박스 예시 코드: 조건문을 사용하여 선택된 투자 성향에 따라 다른 포트폴리오 제공?

## 데이터 입력
- 텍스트 데이터를 입력하고 싶은 경우

    ```st.text_input(value)```

- 텍스트 데이터를 암호로 사용하고 싶은 경우엔 type=”password” 인자를 추가

    ```st.text_input(label, value, type="password")```

- 숫자 데이터를 입력하고 싶은 경우

    ```st.number_input(label, value)```

- 여러 줄의 텍스트 데이터를 입력하고 싶은 경우

    ```st.text_area(label, value)```

- 날짜를 입력하고 싶은 경우

    ```st.date_input(label, value)```

- 시간을 입력하고 싶은 경우

    ```st.time_input(label, value)```
